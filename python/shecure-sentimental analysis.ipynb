```python
#import user reviews dataset CSV file
data=pd.read_csv('Untitled spreadsheet - Sheet1.csv')
#Convert to np array and shuffle the samples
data=np.array(data)
random.shuffle(data)
#Convert to pandas dataframe
data=pd.DataFrame(data)
data
```

```python
#seperate the training and target variables
y_train=data[1]
x_train=data.loc[:,[0]]
x_train.iloc[:,0]
```

```python
#Initialize stopwords and punctuation marks
stopWords = stopwords.words('english')
punct=list(string.punctuation)
lemm = WordNetLemmatizer()
stopWords=stopWords+punct
stopWords.remove('no')
stopWords.remove('nor')
stopWords.remove('not')
stopWords
```

```python
#Function to tokenize the reviews and remove the stopwords and punctuation marks.
def all_words(doc):
    l1=(word_tokenize(doc))
    final_doc=[]
    for l in l1:
        l=''.join([i for i in l if not i.isdigit()])
        for c in string.punctuation:
           l=l.replace(c,"")
        if len(l)>=3:
            if l.lower() not in stopWords:
                final_doc.append(l.lower())
    return(final_doc)
   
```

```python
data=[]
for i in range(x_train.shape[0]):
    sentence=all_words(x_train.iloc[i,0])
    data.append((" ".join(sentence)))
data
```

```python
document=[]
for i in range(y_train.shape[0]):
    document.append((data[i],y_train[i]))
document
```
